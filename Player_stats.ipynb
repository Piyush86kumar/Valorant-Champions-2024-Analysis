```python
percent_cols = ["kast","hs_percent","cl_percent"]
for col in percent_cols:
    df_players[col] = df_players[col].str.replace("%", "").astype(float)
```

```python
df_players = df_players.drop(columns = ["clutches"])
```

```python
sns.histplot(df_players["kills"], bins = 20, kde = True)
plt.title("Distribution of Kills")
plt.show()
```

```python
df_players["kills_per_round"] = (df_players["kills"]/df_players["rounds"]).round(2)
```

```python
df_players.head()
```

```python
sns.histplot(df_players["kills_per_round"],kde = True)
plt.title("Distribution of Kills per round")
plt.show()
```

```python
top_kills = df_players.sort_values(by = "kills", ascending = False).head(10)

sns.barplot(x="kills", y = "player", data = top_kills, hue ="player", palette="Blues_r")
plt.title("Top 10 players by Total kills")
plt.xlabel("Kills")
plt.ylabel("Players")
plt.show()
```

```python
top_kpr = df_players.sort_values(by="kills_per_round", ascending = False,).head(10)

sns.barplot(x = "kills_per_round",y = "player", data = top_kpr, hue = "player", palette="Greens_r")
plt.title("Top Players by Kills per Round")
plt.xlabel("Kills per Round")
plt.ylabel("Players")
plt.show()
```

```python
top_rating = df_players.sort_values(by = "rating", ascending = False).head(10)

sns.barplot(x = "rating", y = "player", data = top_rating, hue = "player", palette= "Reds_r")
plt.title("Top 10 players by rating")
xlabel = "Rating"
ylabel = "Player"
plt.show()
```

```python
top_acs = df_players.sort_values( by = "acs", ascending = False).head(10)

sns.barplot(x = "player", y = "acs", data = top_acs, hue = "player",palette="Blues_r")
plt.title("Top 10 players by ACS")
plt.xlabel("Players")
plt.ylabel("ACS")
plt.show()
```

```python
top_assist = df_players.sort_values(by = "assists", ascending = False).head(10)

sns.barplot(x="assists", y = "player", data = top_assist, hue = "assists")
plt.title("Top 10 players by assist")
plt.xlabel("Assist")
plt.ylabel("Players")
plt.show()
```

```python
# Kills vs Death, (Are fraggers dying more?)

sns.scatterplot(x = "kills", y = "deaths", data = df_players,s = 80, color = "red")
# Add regression line (trend)
sns.regplot(x="kills", y="deaths", data=df_players,scatter=False, color="red", line_kws={"lw":2})
plt.title("Kills vs Deaths")
plt.xlabel("Kills", fontsize = 14)
plt.ylabel("Deaths", fontsize = 14)
plt.show()
```

```python
# Assists vs KD_ratio (do support players still maintain good efficiency?)

df_players["assist_per_rd"] = (df_players["assists"]/df_players["rounds"]).round(2)
sns.regplot(x = "assist_per_rd", y = "kd_ratio", data = df_players)
plt.suptitle("Assist vs Kd")
plt.title("Do support players have to compromise on their kills?")
plt.xlabel("Assist per round")
plt.ylabel("KD Ratio")
plt.show()
```

```python
# Average KD ratio by team (Top 10)

top_KDteam = df_players.groupby("team")["kd_ratio"].mean().sort_values(ascending = False).head(10)
sns.barplot(x= top_KDteam.values, y = top_KDteam.index, hue = top_KDteam.index, palette = "viridis")
plt.title("Average KD ratio by team (Top 10)")
plt.xlabel("Team KD Ratio")
plt.ylabel("Team")
plt.show()
```

```python
# Average vlr rating of team (top 10)

top_rating_team = df_players.groupby("team")["rating"].mean().sort_values(ascending = False).head(10)

sns.barplot(x  = top_rating_team.values, y = top_rating_team.index, hue = top_rating_team.index, palette = "Reds_r")
plt.title("Top 10 Teams based on average vlr rating")
plt.xlabel("Average Team Rating")
plt.ylabel("Teams")
plt.show()
```

```python
# first kill, first death by team 
top_openers = df_players.groupby("team")[["fkpr","fdpr"]].mean().reset_index()  #reset_index() - flatten the index back into a column.

team_openers_melted = top_openers.melt( id_vars = "team", value_vars = ["fkpr","fdpr"], var_name = "type", value_name = "value")

sns.barplot(x = "team", y = "value", hue = "type", data = team_openers_melted)
plt.title("First Kills and First Death percentage for each team")
plt.xlabel("Team")
plt.ylabel("First Kills/ Deaths")
plt.xticks(rotation = 45, ha = "right") # rotates the labels 45° diagonally instead of keeping them flat or horizontal which is default for seaborn
plt.legend(title = "Metric")
plt.show()
```

```python
# top 10 clutch players
top_clutch = df_players.sort_values(by = "cl_percent", ascending = False).head(10)

sns.barplot(x = "cl_percent", y = "player", data = top_clutch, hue = "player", palette = "Blues_r")
plt.title("Top 10 clutch players")
plt.xlabel("Clutch percentage")
plt.ylabel("Players")
plt.show()
```

```python
# top clutch teams of the tournament
team_clutch = df_players.groupby("team")["cl_percent"].mean().sort_values(ascending = False)

sns.barplot(x = team_clutch.index, y = team_clutch.values, hue = team_clutch.index)
plt.title("Team sorted by clutch percentage")
plt.xlabel("Team")
plt.ylabel("Clutch Percentage")
plt.show()
```

```python
# Sort team by hs% . Do team twith higher hs% have more probability of winning
team_hs = df_players.groupby("team")["hs_percent"].mean().sort_values(ascending = False)

sns.barplot(x = team_hs.index, y = team_hs.values, hue = team_hs.index)
plt.title("Teams sorted by headshot percentage")
plt.xlabel("Teams")
plt.ylabel("Headshot percentage")
plt.show()
```

```python
# Ranking players across different parameters like Fk, FD, kills, deaths
weights = {'fk' :0.20,
           'fd' :0.20,
           'kill' : 0.15,
           'death' : 0.15,
           'adr' : 0.15,
           'apr' : 0.15}


df_players["composite_score"] = ( df_players["first_kills"] * weights['fk'] +
                                  df_players["first_deaths"] * weights['fd'] + 
                                  df_players["kills"] * weights['kill'] + 
                                  df_players["deaths"] * weights['death']+
                                  df_players["adr"] * weights['adr'] +
                                  df_players["apr"] * weights['apr']).round(2)

top_com_players = df_players.sort_values(by = 'composite_score', ascending = False).head(15) 

sns.barplot(x = 'composite_score', y = 'player', data = top_com_players, hue = 'player')
plt.title("Composite score of players")
plt.xlabel("Composite Score")
plt.ylabel("Players")
plt.show()
```

```python
# best team as per the combined rating
top_comp_team = df_players.groupby('team')["composite_score"].mean().sort_values(ascending = False).reset_index()
sns.barplot(x = "composite_score", y = "team", data = top_comp_team,hue = "team")
plt.title("Best Teams as per the combined rating of the Players")
plt.xlabel("Composite Score")
plt.ylabel("Teams")
plt.show()
```

```python
# which player has played the most agents using (agents count)
most_agents = df_players.sort_values(by = "agents_count", ascending = False).head(10)

sns.barplot(x = "agents_count", y = "player", data = most_agents, hue = "player", alpha = 1)
plt.title("Most number of agents played by any player")
plt.xlabel("Agent Count")
plt.ylabel("Players")
plt.show()

```

```python
# Corelation Heatmep of performance metrics

metrics_cols = ['rating', 'acs','kd_ratiio', 'adr','kpr','fkpr','fdpr','kast']
num_cols = [c for c in metrics_cols if c in df_players.columns]
corr = df_players[num_cols].corr()
sns.heatmap(corr, annot = True, fmt = '.2f', cmap = 'coolwarm', square = True)
plt.title("Corelation Heatmap of Performance Metrics")
plt.tight_layout() 
plt.show()
```

```python
# PCA

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Select the relevant metrics
stats= df_players[["kills", "deaths","assists","adr","kd_ratio","fkpr","fdpr","rating"]]

# Scale Data 
scaler = StandardScaler()
stats_scaled = scaler.fit_transform(stats)

#Apply PCA
pca = PCA(n_components = 2)
stats_pca = pca.fit_transform(stats_scaled)

# put back into dataframe
pca_df = pd.DataFrame(stats_pca, columns = ["PC1","PC2"])
pca_df["player"] = df_players["player"]
pca_df["team"] = df_players["team"]

# visualize PCA
sns.scatterplot(x = "PC1", y = "PC2", data = pca_df, s = 70, alpha = 1)
plt.title("PCA projection of Players (Performance metrics)")
plt.show()
```

```python
# Distribution overlap between ACS and Kills
from sklearn.preprocessing import MinMaxScaler

# 1. Normalize the data to compare distributions on the same scale
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(
    scaler.fit_transform(df_players[['acs', 'kills']]), 
    columns=['acs_scaled', 'kills_scaled']
)

# 2. Create the KDE (Kernel Density Estimate) plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df_scaled['acs_scaled'], fill=True, label='ACS (Scaled)', color='blue', alpha=0.5)
sns.kdeplot(df_scaled['kills_scaled'], fill=True, label='Kills (Scaled)', color='orange', alpha=0.5)

# 3. Formatting the plot
plt.title('Distribution Overlap: Scaled ACS vs. Total Kills', fontsize=14)
plt.xlabel('Normalized Scale (0 to 1)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
```
